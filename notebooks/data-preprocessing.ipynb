```python
import pandas as pd
import numpy as np
import datetime
```

```python
df = pd.read_csv("../data/raw/movie_data.csv")
```

```python
df = df.drop_duplicates(keep='first')

df.head(5)
```

```python
df.drop(['Metascore', 'Type', 'DVD', 'Production', 'Website', 
         'Response','Season', 'totalSeasons','Episode', 'seriesID', 'Released'], axis=1, inplace=True)
```

```python
df.info()
```

```python
df['Rated'] = df['Rated'].fillna('Unrated')
df['Plot'] = df['Plot'].fillna('')
df['Year'] = df['Year'].fillna('Unknown')
df['Runtime'] = df['Runtime'].fillna('Unknown')
df['Awards'] = df['Awards'].fillna('No award')
df['BoxOffice'] = df['BoxOffice'].fillna('Unknown')
df['Poster'] = df['Poster'].fillna('Not available')
df['Country'] = df['Country'].fillna('Unknown')
df['Language'] = df['Language'].fillna('Unknown')
df['Writer'] = df['Writer'].fillna('Unknown')
df['Director'] = df['Director'].fillna('Unknown')

df['Year'] = pd.to_datetime(df['Year'], format='%Y', errors='coerce')
```

```python
df = df.dropna(subset=['Actors', 'Plot', 'Genre', 'imdbVotes', 'imdbRating'])

df['imdbVotes'] = df['imdbVotes'].replace(',', '', regex=True).fillna(0).astype(int)
```

```python
df.info()
```

```python
rating = df[df['imdbRating'].notnull()]['imdbRating'].astype('int')
C = rating.mean()
C
```

```python
vote_count = df[df['imdbVotes'].notnull()]['imdbVotes'].astype('int')
m = vote_count.quantile(0.9)
m
```

```python
qualified = df[(df['imdbRating'] >= C) & (df['imdbVotes'] >= m)]
qualified.shape
```

```python
def weighted_rating(x):
    v = x['imdbVotes']
    R = x['imdbRating']
    return (v/(v+m) * R) + (m/(m+v) * C)
```

```python
qualified = qualified.copy()

# GÃ¡n giÃ¡ trá»‹ vÃ o cá»™t 'wr'
qualified['wr'] = qualified.apply(weighted_rating, axis=1)

# Sort
qualified = qualified.sort_values('wr', ascending=False)
```

```python
qualified.to_csv('../data/processed/movie_data_processed.csv')
```
