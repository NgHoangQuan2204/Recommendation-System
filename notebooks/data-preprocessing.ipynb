```python
import pandas as pd
import numpy as np
import datetime
```

```python
df = pd.read_csv("../data/raw/full_movie_data.csv", index_col=False)

df.head(5)
```

```python
df.drop(['Metascore', 'Type', 'DVD', 'Production', 'Website', 
         'Response','Season', 'totalSeasons','Episode', 'seriesID', 'Released'], axis=1, inplace=True)
```

```python
df.info()
```

```python
df['Rated'] = df['Rated'].fillna('Unrated')
df['Plot'] = df['Plot'].fillna('')
df['Year'] = df['Year'].fillna('Unknown')
df['Runtime'] = df['Runtime'].fillna('Unknown')
df['Awards'] = df['Awards'].fillna('No award')
df['BoxOffice'] = df['BoxOffice'].fillna('Unknown')
df['Poster'] = df['Poster'].fillna('Not available')
df['Country'] = df['Country'].fillna('Unknown')
df['Language'] = df['Language'].fillna('Unknown')
df['Writer'] = df['Writer'].fillna('Unknown')
df['Director'] = df['Director'].fillna('Unknown')

df['Year'] = pd.to_datetime(df['Year'], format='%Y', errors='coerce')
```

```python
df = df.dropna(subset=['Actors', 'Plot', 'Genre', 'imdbVotes', 'imdbRating'])

df['imdbVotes'] = df['imdbVotes'].replace(',', '', regex=True).fillna(0).astype(int)
```

```python
df.info()
```

```python
rating = df[df['imdbRating'].notnull()]['imdbRating'].astype('int')
C = rating.mean()
C
```

```python
vote_count = df[df['imdbVotes'].notnull()]['imdbVotes'].astype('int')
m = vote_count.quantile(0.9)
m
```

```python
qualified = df[(df['imdbRating'] >= C) & (df['imdbVotes'] >= m)]
qualified.shape
```

```python
def weighted_rating(x):
    v = x['imdbVotes']
    R = x['imdbRating']
    return (v/(v+m) * R) + (m/(m+v) * C)
```

```python
qualified = qualified.copy()

# GÃ¡n giÃ¡ trá»‹ vÃ o cá»™t 'wr'
qualified['wr'] = qualified.apply(weighted_rating, axis=1)

# Sort
qualified = qualified.sort_values('wr', ascending=False)
```

```python
qualified.info()
```

```python
qualified = qualified.drop_duplicates('Title', ignore_index= True)

qualified.head(5)
```

```python
qualified.to_csv('../data/processed/movie_data_processed.csv', index= False)
```

```python
ratings_df = pd.read_csv('../data/raw/ratings.csv', index_col=False)
```

```python
ratings_df.dropna()
```

```python
ratings_df.head(5)
```

```python
links_df = links_df.rename(columns= {'imdbId': 'imdbID'})
```

```python
movie_rating_counts = ratings_df.groupby('movieId')['userId'].count()

popular_movies = movie_rating_counts[movie_rating_counts > 5000].index
del movie_rating_counts

df_ratings_filtered = ratings_df[ratings_df['movieId'].isin(popular_movies)]
del popular_movies

df_ratings_filtered.info()
```

```python
start = 0
chunk_size = 1000000

merged_ratings_df = pd.DataFrame()

while True:
    chunk = df_ratings_filtered.iloc[start:start + chunk_size]

    if chunk.empty:
        break
    
    merged_ratings_df = merged_ratings_df._append(chunk.merge(links_df, on="movieId", how="inner"), ignore_index=True)
    
    start += chunk_size

results = []

for chunk in range(0, len(merged_ratings_df), chunk_size):
    merged_ratings_df_chunk = merged_ratings_df.iloc[chunk:chunk + chunk_size]
    result_chunk = merged_ratings_df_chunk['imdbID'].apply(lambda x: f'tt{str(x).zfill(7)}')
    results.append(result_chunk)

merged_ratings_df['imdbID'] = pd.concat(results)

merged_ratings_df.head(5)
```

```python
ratings_df = merged_ratings_df[merged_ratings_df['imdbID'].isin(qualified['imdbID'])]
ratings_df = ratings_df.drop(columns='movieId')
ratings_df.head(5)
```

```python
ratings_df.info()
```

```python
ratings_df.to_csv("../data/processed/ratings_processed.csv", index= False)
```
